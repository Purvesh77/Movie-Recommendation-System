```python
from tmdbv3api import Movie
import requests
import numpy as np

tmdb_movie = Movie()

def get_genre(title):
    genres = []
    try:
        results = tmdb_movie.search(title)
       
        # Print the search results for debugging
       

        # Find the first result that is an instance of the Movie class
        movie_result = results[0]

        # Print the selected movie result for debugging

        if movie_result:
            movie_id = movie_result.id
            response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb_movie.api_key}')
            data_json = response.json()

            if data_json.get('genres'):
                genres = [genre['name'] for genre in data_json['genres']]
                # print(genres)
                genre_str = " ".join(genres)
                # print(f"{title}: {genre_str}")  # Print the title and genre
                return genre_str
        else:
            print(f"Skipping '{title}' - Not a movie or no results found")
    except Exception as e:
        print(f"Error processing '{title}': {e}")
    return np.NaN

# Assuming df is your DataFrame
df['genres'] = df['Title'].apply(get_genre)
```

```python
def get_director(x):
    if " (director)" in x:
        return x.split(" (director)")[0]
    elif " (directors)" in x:
        return x.split(" (directors)")[0]
    else:
        return x.split(" (director/screenplay)")[0]
```

```python
df_2018['director_name'] = df_2018['Cast and crew'].map(lambda x: get_director(x))
```

```python
def get_actor1(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
df_2018['actor_1_name'] = df_2018['Cast and crew'].map(lambda x: get_actor1(x))
```

```python
def get_actor2(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
df_2018['actor_2_name'] = df_2018['Cast and crew'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
df_2018['actor_3_name'] = df_2018['Cast and crew'].map(lambda x: get_actor3(x))
```
